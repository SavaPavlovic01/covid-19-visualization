```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv("covid19_Confirmed_dataset.csv")
df.head()
```

```python
df.shape
```

```python
df.drop(["Lat", "Long", "Province/State"], axis = 1, inplace = True)
```

```python
df.head()
```

```python
aggregation = df.groupby("Country/Region").sum()
```

```python
aggregation.head()
```

```python
aggregation.shape
```

```python
aggregation.loc["China"].plot()
aggregation.loc["Italy"].plot()
aggregation.loc["Spain"].plot()
plt.legend()
```

```python
aggregation.loc["China"].plot()
```

```python
aggregation.loc["China"][:3].plot()
```

```python
aggregation.loc["China"].diff().plot()
```

```python
aggregation.loc["China"].diff().max()
```

```python
aggregation.loc["Italy"].diff().max()
```

```python
aggregation.loc["Spain"].diff().max()
```

```python
countries=list(aggregation.index)
max_infection_rates=[]
for c in countries:
    max_infection_rates.append(aggregation.loc[c].diff().max())
aggregation["max_infection_rates"]=max_infection_rates
```

```python
aggregation.head()
```

```python
data = pd.DataFrame(aggregation["max_infection_rates"])
```

```python
data.head()
```

```python
happiness = pd.read_csv("worldwide_happiness_report.csv")
```

```python
happiness.head()
```

```python
cols = ["Overall rank","Score","Generosity","Perceptions of corruption"]
```

```python
happiness.drop(cols,axis=1,inplace=True)
happiness.head()
```

```python
happiness.set_index("Country or region",inplace=True)
happiness.head()
```

```python
final = data.join(happiness,how="inner")
final.head()
```

```python
final.corr()
```

```python
x = final["GDP per capita"]
y = final["max_infection_rates"]
sns.scatterplot(x = x,y = np.log(y))
```

```python
sns.regplot(x = x, y = np.log(y))
```
